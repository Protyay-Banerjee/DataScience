# how many Null value count 

```python
#isnull function 
data.isnull().sum()
plt.figure(figsize=(32,12))
sns.heatmap(data.isnull(),yticklabels=True,cbar=True,cmap='viridis')
plt.show()
```

# percentage find Null value 

```python
x=data.isnull().sum()/data.shape[0] * 100 # data.shape[0] is all row value 
print(x)



```

# perticular condition wise value 

```python
drop_column=x[x>10].keys()
print(drop_column)              
```

```python
#drop columns 
data2=data.drop(columns=drop_column)
data2.shape
```

```python
#isnull function 
data2.isnull().sum()
plt.figure(figsize=(32,12))
sns.heatmap(data2.isnull(),yticklabels=True,cbar=True,cmap='viridis')
plt.show()
```

```python
# drop null  row  
data3=data2.dropna()
data3.shape
```

```python
#Finally clean our data
plt.figure(figsize=(32,12))
sns.heatmap(data3.isnull(),yticklabels=True,cbar=True,cmap='viridis')
plt.show()
```
